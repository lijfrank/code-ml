```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np 
points = np.genfromtxt("challenge_dataset.txt", delimiter=",")
X = points[:,0]
Y = points[:,1]
# print(points)
# print(x)
plt.scatter(X, Y)
plt.show()
```

## 损失函数

```python
def compter_cost(w, b, points):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i,0]
        y = points[i,1]
        total_cost += (y-w*x-b)**2
    return total_cost/M
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)
lr.fit(X, Y)
w = lr.coef_[0][0]
b = lr.intercept_[0]
```

```python
print("w is",w)
print("b is",b)
cost = compter_cost(w, b, points)
print(cost)
```

## 画出拟合图

```python
plt.scatter(X, Y)
pre_Y = w * X + b
plt.plot(X, pre_Y, c="r")
```
