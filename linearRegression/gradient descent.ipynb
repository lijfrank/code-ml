```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np 
points = np.genfromtxt("challenge_dataset.txt", delimiter=",")
X = points[:,0]
Y = points[:,1]
# print(points)
# print(x)
plt.scatter(X, Y)
plt.show()
```

## 损失函数

```python
def compter_cost(w, b, points):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i,0]
        y = points[i,1]
        total_cost += (y-w*x-b)**2
    return total_cost/M
```

## 定义超参数

```python
alpha = 0.001
initial_w = 0
initial_b = 0
num_iter = 50
```

## 定义梯度下降算法

```python
def grad_desc(points, initial_w, initial_b, num_iter, alpha):
    w = initial_w
    b = initial_b
    cost_list = []
    for i in range(num_iter):
        cost_list.append(compter_cost(w,b,points))
        w ,b = step_grad_desc(w,b,alpha,points)
    return [w,b,cost_list]
def step_grad_desc(current_w,current_b,alpha,points):
    sum_grad_w = 0
    sum_grad_b = 0
    M = len(points)
    for i in range(M):
        x = points[i, 0]
        y = points[i, 1]
        sum_grad_w += (current_w*x-current_b-y)*x
        sum_grad_b += current_w*x-current_b-y
    grad_w = 2/M *sum_grad_w
    grad_b = 2/M *sum_grad_b
    updated_w = current_w - alpha*grad_w
    updated_b = current_b - alpha*grad_b
    return updated_w,updated_b
```

```python
w, b, cost_list = grad_desc(points, initial_w, initial_b,num_iter,alpha)
print("w is",w)
print("b is",b)
plt.plot(cost_list)###索引值为x，值为y画图
cost = compter_cost(w, b, points)
print(cost)
```
